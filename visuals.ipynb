```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
df = pd.read_csv("D:/AI WORKSHOP/TASK/BOOSTING_task/UCI_Credit_Card.csv")
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="default.payment.next.month", data=df, palette="viridis")
plt.title("Default vs No Default Distribution")
plt.xlabel("Default (1 = Yes, 0 = No)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = df.corr()
sns.heatmap(corr, cmap="coolwarm", annot=False, cbar=True)
plt.title("Correlation Heatmap of Features")
plt.show()
```

```python
cont_features = ["LIMIT_BAL", "AGE", "BILL_AMT1", "PAY_AMT1"]

plt.figure(figsize=(15,8))
for i, col in enumerate(cont_features, 1):
    plt.subplot(2,2,i)
    sns.histplot(df[col], kde=True, bins=30, color="blue")
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```
